```python
#expansion
gdp_data = pd.read_csv("C:/Users/rrose/Documents/project 1/GDP.csv")
    
```

```python
gdp_data = gdp_data.sort_values(by="DATE")
```

```python
gdp_data["quarterly_diff"]= gdp_data["GDP"].diff().shift(1)
```

```python
gdp_data["percentage_change"]= gdp_data["quarterly_diff"].pct_change()*100
```

```python
gdp_data.head()
```

```python
gdp_data = gdp_data.dropna()
```

```python
gdp_data.head()
```

```python

gdp_data["DATE"]= gdp_data["DATE"].astype("category")
gdp_data.hvplot.bar(    
    x='DATE',
    y='percentage_change',
    title='GDP Percent Change',rot=90
)

```
